```python
data['win_by_runs'] = data['win_by_runs'].fillna(data['win_by_runs'].mean())
data['win_by_runs'] = data['win_by_runs'].astype(int)
data['win_by_wickets'] = data['win_by_wickets'].fillna(data['win_by_wickets'].mean())
data['win_by_wickets'] = data['win_by_wickets'].astype(int)
data['second_bat_score'] = data['second_bat_score'].fillna(data['second_bat_score'].mean())
data['second_bat_score'] = data['second_bat_score'].astype(int)
```

```python
data['dl_applied'] = data['dl_applied'].replace(np.nan,'NoD/L')
```

```python
data['eliminator'] = data['eliminator'].fillna(data['eliminator'].mode()[0])
data['city'] = data['city'].fillna(data['city'].mode()[0])
data['winner'] = data['winner'].fillna(data['winner'].mode()[0])
data['result'] = data['result'].fillna(data['result'].mode()[0])
data['player_of_match'] = data['player_of_match'].fillna(data['player_of_match'].mode()[0])
data['umpire3'] = data['umpire3'].fillna(data['umpire3'].mode()[0])
data['umpire2'] = data['umpire2'].fillna(data['umpire2'].mode()[0])
data['umpire1'] = data['umpire1'].fillna(data['umpire1'].mode()[0])
```
